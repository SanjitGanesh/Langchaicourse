#### Messages
###### Messages are the fundamental unit of context for models in LangChain. They represent the input and output of models, carrying both the content and metadata needed to represent the state of a conversation when interacting with an LLM. Messages are objects that contain:

###### Role - Identifies the message type (e.g. system, user)
###### Content - Represents the actual content of the message (like text, images, audio, documents, etc.)
###### Metadata - Optional fields such as response information, message IDs, and token usage
###### LangChain provides a standard message type that works across all model providers, ensuring consistent behavior regardless of the model being called.

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:qwen/qwen3-32b")
```

##### Text prompts
###### text prompts are strings - ideal for straightforward generation tasks where you don't need to retain the conversational history

```python
model.invoke("what is honeytrap?")
```

Messages

```python
from langchain.messages import SystemMessage,HumanMessage,AIMessage

messages = [
    SystemMessage("you are a poetry expert"),
    HumanMessage("write a poem on moon")
]

response = model.invoke(messages)
response.content
```

```python
system_msg = SystemMessage("you are a helpful coding assistance")
messages = [
    system_msg,
    HumanMessage("How do I create a REST API?")
]

response = model.invoke(messages)
print(response.content)
```

```python
## Detailed info to the LLM through system message
from langchain.messages import HumanMessage,SystemMessage

system_msg = SystemMessage("""
your'e a senior python developer with the expertise in webframeworks.
Always provide code examples and explain your reasoning.
Be concise but thorough in your explainations.
""")

messages = [
    system_msg,
    HumanMessage("How do I create a REST API?")
]
response = model.invoke(messages)
print(response.content)
```

```python
## Human Metadata
human_msg = HumanMessage(
    content="Hello!",
    name = "alice", # identify different users - optional
    id = "msg_123", # Unique identifier for tracing - optional
)
```

```python
response = model.invoke([
    human_msg
])
response
```

```python
from langchain.messages import AIMessage, SystemMessage, HumanMessage

#create a AI message manually (eg for conversational history)
ai_msg = AIMessage("I'd be happy to help you with the question!")

#add conversational history manually 
messages = [
    SystemMessage("your'e a helpfull assistant"),
    HumanMessage("Can you help me?"),
    ai_msg,
    HumanMessage("Great! What's 2+2?")
]

response = model.invoke(messages)
print(response.content)
```

```python
from langchain.messages import AIMessage
from langchain.messages import ToolMessage

# After a model makes a tool call
# (Here, we demonstrate manually creating the messages for brevity)
ai_message = AIMessage(
    content=[],
    tool_calls=[{
        "name": "get_weather",
        "args": {"location": "Chennai"},
        "id": "call_123"
    }]
)

weather_result = "Sunny, 72°F"
tool_message = ToolMessage(
    content=weather_result,
    tool_call_id="call_123"  # Must match the call ID
)

# Continue conversation
messages = [
    HumanMessage("What's the weather in San Francisco?"),
    ai_message,  # Model's tool call
    tool_message,  # Tool execution result
]
response = model.invoke(messages)  # Model processes the result
```

```python
print(response.content)
```
