### Model integration with google and grog APIs

#### Google Gemini model

```python
import os
from dotenv import load_dotenv

load_dotenv()
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("gemini-2.5-flash",model_provider="google_genai")
response = model.invoke("Why do parrots talk?")
response.content
```

```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model='gemini-2.5-flash',model_provider="google_genai")
response = model.invoke("Why do humans speak?")
response.content
```

### GROQ model integration

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("why do parrots speak?")
response.content
```

```python
import os
from langchain_groq import ChatGroq

model = ChatGroq(model ="qwen/qwen3-32b")
response = model.invoke("why do humans speak?")
response.content
```

### Streaming

##### A model can stream the output while the output is beeen generated, by displaying the output progresively

```python
for chunck in model.stream("why do parrots speak?"):
    print(chunck.text,end="|",flush=True)
```

### Batch

#### batching is the collection of individual inputs and prcessing all of them in parallel, this improves the performance and reduces the cost

```python
responses = model.batch(["Why do parrots have colourful feathers?",
"who found the aeroplane?",
"what is unreal engine?"])
for response in responses:
    print(response)
```

```python
responses = model.batch(["Why do parrots have colourful feathers?",
"who found the aeroplane?",
"what is unreal engine?"],
config = {
    'max_concurrency':5 #limit the no of parrallel calls, if 10: 5/5
}
)
for response in responses:
    print(response)
```
